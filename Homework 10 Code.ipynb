```python
G = pp.graphs.ErdosRenyi(N = 50,p = 0.1)
GR = pp.graphs.Ring(N=50)

G.set_coordinates(kind='spring', seed=42)
fig, axes = plt.subplots(1, 2)
_ = axes[0].spy(G.W, markersize=2)
G.plot(ax=axes[1])

GR.set_coordinates(kind='spring', seed=42)
fig, axes = plt.subplots(1, 2)
_ = axes[0].spy(G.W, markersize=2)
GR.plot(ax=axes[1])
```
