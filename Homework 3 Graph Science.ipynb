```python
import networkx as nx
import numpy as np
import math as math
import matplotlib.pyplot as plt


g1 = nx.Graph()
g1.add_nodes_from([2,5,7])
g1.add_edges_from([(5,2),(7,2),(5,7)])

g1.add_nodes_from([3,4,6,1])
g1.add_edges_from([(3,4),(3,6),(4,6),(4,1),(6,1)])
nx.draw(g1)
```

```python
L = nx.laplacian_matrix(g1)
print(L)
```

```python
IM = nx.incidence_matrix(g1)
TIM = np.transpose(g1)
print(IM)
print(TIM)

```
