```python
def draw(G, measures, measure_name):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1, base=10))
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
```

```python
D=nx.DiGraph()
Dnodes= G.nodes
Dedges= G.edges
D.add_nodes_from(Dnodes)
D.add_edges_from(Dedges)

pos = nx.spring_layout(D, seed=675)
draw(D, nx.degree_centrality(D), 'Degree Centrality') 
#draw(D, pos, nx.eigenvector_centrality(D), 'Eigenvector Centrality') 

Adj_D=nx.adjacency_matrix(D);
```
