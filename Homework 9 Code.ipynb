```python
import numpy as np
import networkx as nx
import pandas as pd
from node2vec import Node2Vec
from numpy import linalg as LA
import scipy as sp
import sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import geopandas
from sklearn.cluster import KMeans
```

# Part a

```python
streets = pd.read_csv("/Users/jacobeverly/Documents/Graph Science/HW9/streets.csv")
nodes = pd.read_csv("/Users/jacobeverly/Documents/Graph Science/HW9/nodes.csv")
```

```python
G = nx.Graph() 
for i in range(streets.shape[0]):
    G.add_edge(streets['u'][i], streets['v'][i], weight = streets['length'][i])
for i in range(nodes.shape[0]):
        G.add_node(nodes['osmid'][i], pos = (nodes['y'][i],nodes['x'][i]))
```

# Part b

```python
G.remove_edges_from(nx.selfloop_edges(G))
```

```python
pos = nx.get_node_attributes(G, "pos")
```

```python
nx.draw(G,pos,node_size = 1, node_color = 'red')
```

# Part c

```python
node2vec = Node2Vec(G, dimensions=32, walk_length=50, num_walks=300)
```

```python
model = node2vec.fit(window=10, min_count=1, batch_words=4)
```

# Part d

```python
laplaplaplaplap = nx.laplacian_matrix(G, weight = 'weight')
laplaplaplaplap = laplaplaplaplap.todense()
w, v = LA.eig((laplaplaplaplap))
```

```python
from node2vec.edges import WeightedL1Embedder
edges_embs = WeightedL1Embedder(keyed_vectors=model.wv)
```

# Part e

# Laplacian Graph

```python
g_nx_ccs = (G.subgraph(c).copy() for c in nx.connected_components(G))
g_nx = max(g_nx_ccs, key=len)
node_ids = list(G.nodes())
node_targets = [ G.nodes[node_id] for node_id in node_ids]

tsne = TSNE()
L_pr = tsne.fit_transform(laplaplaplaplap)
```

```python
alpha=0.7
fig = plt.figure(figsize=(10,8))
plt.scatter(L_pr[:,0],L_pr[:,1], alpha=alpha)
```

# Node2Vec Graph

```python
g_nx_ccs = (G.subgraph(c).copy() for c in nx.connected_components(G))
g_nx = max(g_nx_ccs, key=len)
node_ids = list(G.nodes())
node_targets = [ G.nodes[node_id] for node_id in node_ids]
```

```python
embs = []
for u,v,a in G.edges(data=True):
    x = edges_embs[(str(u),str(v))]
    embs.append(x)
```

```python
tsne = TSNE()
v_pr = tsne.fit_transform(embs)
fig = plt.figure(figsize=(10,8))
plt.scatter(v_pr[:,0],v_pr[:,1], cmap="jet")
```

# Laplacian

```python
kmeans = KMeans(n_clusters=32)
kmeans.fit(L_pr)
y_kmeans = kmeans.predict(L_pr)
```

```python
plt.scatter(L_pr[:, 0], L_pr[:, 1], c=y_kmeans, s=50)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```
